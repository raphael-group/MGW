```python
import os, json, pathlib as pl
import numpy as np
import pandas as pd
import scipy.sparse as sp
import anndata as ad
import scanpy as sc


# ---- inputs (same as shell above; change here if running standalone) ----

BASE="/scratch/gpfs/ph3641/mgw/sma_data/mendeley_all/Spatial Multimodal Analysis of Transcriptomes and Metabolomes in Tissues/processed/"

out_dir = pl.Path("/scratch/gpfs/ph3641/mgw/sma_data/")
out_dir.mkdir(parents=True, exist_ok=True)

#ST_OUTS = f"{BASE}/visium/visium/V11T17-101/V11T17-101_A1/output_data/V11T17-101_A1/outs"
#MSI_CSV = f"{BASE}/msi/msi/ITO1/ITO1_A1/output_data/ITO1.mPD1.9aa.220826.csv"
ST_OUTS = "/scratch/gpfs/ph3641/mgw/sma_data/mendeley_all/Spatial Multimodal Analysis of Transcriptomes and Metabolomes in Tissues/processed/sma/sma/V11T17-102/V11T17-102_A1/output_data/V11T17-102_A1_RNA/outs"
MSI_CSV = "/scratch/gpfs/ph3641/mgw/sma_data/mendeley_all/Spatial Multimodal Analysis of Transcriptomes and Metabolomes in Tissues/processed/sma/sma/V11T17-102/V11T17-102_A1/output_data/V11T17-102_A1_MSI/V11T17-102_A1.Visium.FMP.220826_smamsi.csv"
META_CSV="/scratch/gpfs/ph3641/mgw/sma_data/mendeley_all/Spatial Multimodal Analysis of Transcriptomes and Metabolomes in Tissues/metadata.csv"

# =========================
# A) VISIUM (ST)
# =========================
h5 = pl.Path(ST_OUTS) / "filtered_feature_bc_matrix.h5"
spatial_dir = pl.Path(ST_OUTS) / "spatial"

adata_st = sc.read_10x_h5(h5)  # barcodes in obs_names, genes in var_names
adata_st.var_names_make_unique()

# ---- spatial spots
# 10x has two possible files: tissue_positions_list.csv OR tissue_positions.csv
tpl = None
for fname in ["tissue_positions_list.csv", "tissue_positions.csv"]:
    f = spatial_dir / fname
    if f.exists():
        tpl = pd.read_csv(f, header=None)
        break
if tpl is None:
    raise FileNotFoundError("Could not find tissue_positions_*.csv in spatial/")

# Normalize columns per 10x conventions
# v1 (list): barcode, in_tissue, array_row, array_col, px_row, px_col
# v2 (csv with header) sometimes similar; we handle headerless by position
if tpl.shape[1] >= 6 and tpl.columns.dtype == 'int64':
    tpl.columns = ["barcode","in_tissue","array_row","array_col","px_col","px_row"] if tpl.shape[1]==6 else \
                  ["barcode","in_tissue","array_row","array_col","px_col","px_row"] + [f"extra_{i}" for i in range(tpl.shape[1]-6)]
else:
    # If there's a header, rename to a standard set
    cols_lower = [str(c).lower() for c in tpl.columns]
    # try to locate required fields
    needed = ["barcode","in_tissue","array_row","array_col","px_col","px_row"]
    mapping = {}
    for want in needed:
        if want in cols_lower:
            mapping[tpl.columns[cols_lower.index(want)]] = want
    tpl = tpl.rename(columns=mapping)
    missing = [c for c in ["barcode","px_col","px_row"] if c not in tpl.columns]
    if missing:
        raise ValueError(f"Missing columns in tissue positions: {missing}")

tpl = tpl.set_index("barcode")
tpl = tpl.reindex(adata_st.obs_names)
adata_st.obs["in_tissue"] = tpl["in_tissue"].astype(int)
adata_st.obsm["spatial"] = tpl[["px_col","px_row"]].to_numpy().astype(float)

# ---- scalefactors & images
sf_json = spatial_dir / "scalefactors_json.json"
if sf_json.exists():
    scalefactors = json.load(open(sf_json))
else:
    scalefactors = {}
adata_st.uns["spatial"] = {
    "hires_image": str(spatial_dir / "tissue_hires_image.png"),
    "lowres_image": str(spatial_dir / "tissue_lowres_image.png"),
    "scalefactors": scalefactors
}

# ---- region / lesion labels (if present)
for extra_csv in ["region.csv", "RegionLoupe.csv", "lesion.csv"]:
    f = pl.Path(ST_OUTS) / extra_csv
    if f.exists():
        df = pd.read_csv(f)
        # Try to join by barcode (common in these files)
        # Look for barcode column variants
        key = None
        for cand in ["barcode","Barcode","spot","Spot","barcode_id","Spot.Barcode"]:
            if cand in df.columns:
                key = cand
                break
        if key is not None:
            df = df.set_index(key).reindex(adata_st.obs_names)
            for c in df.columns:
                if c.lower() not in {"x","y","row","col"}:
                    adata_st.obs[c] = df[c].astype("category") if df[c].dtype == object else df[c]

# Provenance
adata_st.uns["provenance"] = {
    "source": "Mendeley visium.zip",
    "outs_dir": str(pl.Path(ST_OUTS))
}

# =========================
# B) MSI
# =========================
# The CSV may be "wide" (pixels as rows; m/z bins as columns) OR "long" (x,y,mz,intensity rows).
# We'll auto-detect:

df_head = pd.read_csv(MSI_CSV, nrows=50)
cols = [c.lower() for c in df_head.columns]

def load_msi_wide(path):
    df = pd.read_csv(path)
    # try to detect coordinate columns
    possible_x = [c for c in df.columns if str(c).lower() in {"x","col","column","px","i","ix"}]
    possible_y = [c for c in df.columns if str(c).lower() in {"y","row","py","j","iy"}]
    # Heuristic: if first two columns are coords
    coord_cols = []
    if len(df.columns) >= 3:
        # if dtypes of first two look integer-ish, assume coords
        first_two = df.iloc[:, :2]
        if np.issubdtype(first_two.dtypes[0], np.integer) and np.issubdtype(first_two.dtypes[1], np.integer):
            coord_cols = list(df.columns[:2])
    if not coord_cols:
        # fallback to named
        xcol = possible_x[0] if possible_x else df.columns[0]
        ycol = possible_y[0] if possible_y else df.columns[1]
        coord_cols = [xcol, ycol]

    # build obs from coords, remaining columns are m/z bins
    obs = df[coord_cols].copy()
    obs.columns = ["x","y"]
    mat = df.drop(columns=coord_cols)
    # try to coerce columns to mz floats if possible
    try:
        mz = np.array([float(str(c).replace("mz","").replace("m/z","").replace("_","").replace(" ","")) for c in mat.columns])
    except Exception:
        mz = np.arange(mat.shape[1], dtype=float)

    X = sp.csr_matrix(mat.values.astype(np.float32))
    adata_msi = ad.AnnData(X=X, obs=obs, var=pd.DataFrame({"mz": mz}))
    adata_msi.obsm["spatial"] = obs[["x","y"]].to_numpy().astype(np.float32)
    return adata_msi

def load_msi_long(path):
    # expect columns like: x,y,mz,intensity  (order may vary)
    usecols = None
    df = pd.read_csv(path)
    lower = {c.lower(): c for c in df.columns}
    for needed in [["x","y","mz","intensity"], ["col","row","mz","intensity"]]:
        if all(k in lower for k in needed):
            usecols = [lower[k] for k in needed]
            break
    if usecols is None:
        raise ValueError("Could not find (x,y,mz,intensity) columns in long MSI CSV")

    xcol,ycol,mzcol,Icol = usecols
    # build pixel id
    df["pid"] = list(zip(df[xcol].astype(int), df[ycol].astype(int)))
    # stable pixel index mapping
    pids = pd.Index(df["pid"].unique(), name="pid")
    pid_to_row = {pid:i for i,pid in enumerate(pids)}

    # m/z centers
    mz_vals = np.asarray(df[mzcol].values, dtype=float)
    mz_centers, inv = np.unique(mz_vals, return_inverse=True)
    n_pix = len(pids); n_mz = len(mz_centers)

    rows = df["pid"].map(pid_to_row).to_numpy()
    cols = inv
    vals = df[Icol].to_numpy(dtype=np.float32)

    X = sp.coo_matrix((vals, (rows, cols)), shape=(n_pix, n_mz)).tocsr()
    obs = pd.DataFrame({"x":[pid[0] for pid in pids], "y":[pid[1] for pid in pids]})
    var = pd.DataFrame({"mz": mz_centers})

    adata_msi = ad.AnnData(X=X, obs=obs, var=var)
    adata_msi.obsm["spatial"] = obs[["x","y"]].to_numpy().astype(np.float32)
    return adata_msi

if {"mz","intensity"} & set(cols):
    # likely long format
    adata_msi = load_msi_long(MSI_CSV)
else:
    adata_msi = load_msi_wide(MSI_CSV)

# add provenance / chemistry from filename if useful
msi_path = pl.Path(MSI_CSV)
adata_msi.uns["provenance"] = {
    "source": "Mendeley msi.zip",
    "file": str(msi_path.name),
}
# try to parse matrix chemistry token (e.g., '9aa', 'dhb', 'fmp10') from filename
name_lower = msi_path.name.lower()
for token in ["9aa","dhb","fmp","fmp10","ito1","ito2","ito3","mpd1"]:
    if token in name_lower:
        adata_msi.uns["msi_matrix"] = token

# =========================
# Save both
# =========================
adata_st.write(out_dir/"adata_st_V11T17-102_A1.h5ad", compression="gzip")
adata_msi.write(out_dir/"adata_msi_V11T17-102_A1.h5ad", compression="gzip")

print("Wrote:")
print(out_dir/"adata_st_V11T17-102_A1.h5ad")
print(out_dir/"adata_msi_V11T17-102_A1.h5ad")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# st := your Visium AnnData
xy = st.obsm["spatial"]

# 1) normalize labels to strings and handle missing
lab = st.obs.get("dopamine")
if lab is None:
    raise ValueError("Column 'dopamine' not found in st.obs")
lab = lab.astype(str).replace({"nan": np.nan}).fillna("unknown")

# 2) map every possible category
palette = {
    "CI": "red",
    "dopamine_Cd": "orange",
    "not_dopamine_Cd": "lightgray",
    "unknown": "lightgray"
}
colors = lab.map(palette)
# guard: anything unmapped -> lightgray
colors = colors.fillna("lightgray").to_numpy()

# 3) plot
plt.figure(figsize=(6,6))
plt.scatter(xy[:,0], xy[:,1], c=colors, s=10)
plt.gca().invert_yaxis()
plt.axis("equal")
plt.title("Visium dopamine annotations (CI=red)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.sparse as sp

def to_1d_numeric(col):
    """Return a dense 1-D float array from a variety of sparse/ndarray slices."""
    # common case: scipy sparse matrix (n,1)
    if sp.issparse(col):
        return col.toarray().ravel().astype(float)
    # sometimes you get a length-1 object array that holds a sparse matrix
    if isinstance(col, np.ndarray) and col.dtype == object and col.size == 1 and sp.issparse(col.item()):
        return col.item().toarray().ravel().astype(float)
    # plain numpy
    col = np.asarray(col)
    # (n,1) -> (n,)
    if col.ndim > 1:
        col = col.reshape(col.shape[0], -1)[:, 0]
    return col.astype(float)

# --- pick dopamine-related targets for FMP-10 datasets ---
targets = {
    "DA_single": 421.19,     # dopamine (FMP-10 single deriv.)
    "DA_double": 674.28,     # dopamine (FMP-10 double deriv.)
    "3MT": 435.21,           # 3-methoxytyramine (FMP-10)
}

mz = msi.var["mz"].values.astype(float)

def closest_idx(target, tol_ppm=5):
    idx = int(np.argmin(np.abs(mz - target)))
    # optional sanity check: ensure within tolerance
    ppm = 1e6 * abs(mz[idx] - target) / target
    if ppm > tol_ppm:
        print(f"Warning: closest to {target} m/z is {mz[idx]:.5f} (={ppm:.1f} ppm > {tol_ppm} ppm)")
    return idx

xy = msi.obsm["spatial"]

for label, target in targets.items():
    j = closest_idx(target, tol_ppm=15)  # widen ppm a bit just for visualization
    intens = to_1d_numeric(msi.X[:, j])
    plt.figure(figsize=(6,6))
    plt.scatter(xy[:,0], xy[:,1], c=intens, s=8, cmap="inferno")
    plt.gca().invert_yaxis()
    plt.axis("equal")
    plt.title(f"MSI {label} (target {target:.2f}) matched m/z={mz[j]:.5f}")
    plt.show()
```
