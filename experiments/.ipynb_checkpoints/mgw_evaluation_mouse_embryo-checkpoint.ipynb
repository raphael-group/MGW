```python

# ============================================================
# M-GW: Consolidated Pairwise Evaluation (Mouse Embryo)
# ============================================================
import os, random, json, warnings
import numpy as np
import pandas as pd
import torch
import scanpy as sc
import anndata as ad
import scipy.sparse as sp
from scipy.spatial.distance import cdist
from datetime import datetime
from typing import Dict, Tuple

import matplotlib.pyplot as plt

from mgw import util, plotting, metrics
from mgw import mgw as mgw_mod
from mgw.gw import solve_gw_ott

# ----------------------------
# Configuration
# ----------------------------
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
torch.set_default_dtype(torch.float64)
GLOBAL_SEED = 42

# Input files & params
FILEHANDLES = [
    '/scratch/gpfs/ph3641/mouse_embryo/E9.5_E1S1.MOSTA.h5ad',
    '/scratch/gpfs/ph3641/mouse_embryo/E10.5_E1S1.MOSTA.h5ad',
    '/scratch/gpfs/ph3641/mouse_embryo/E11.5_E1S1.MOSTA.h5ad',
    '/scratch/gpfs/ph3641/mouse_embryo/E12.5_E1S1.MOSTA.h5ad',
    '/scratch/gpfs/ph3641/mouse_embryo/E13.5_E1S1.MOSTA.h5ad'
]

TIMEPOINTS = ['E9.5', 'E10.5', 'E11.5', 'E12.5', 'E13.5']

PCA_COMP = 30
CCA_COMPS = [3, 10, 10, 10]            # per adjacent pair
KNN_K = 12
GEODESIC_EPS = 1e-2
N_downsample = 10_000

GW_PARAMS = dict(verbose=True, inner_maxit=3000, outer_maxit=3000,
                 inner_tol=1e-8, outer_tol=1e-8, epsilon=1e-4)

# Directories
ROOT_SAVE = "/scratch/gpfs/ph3641/mgw/ME_Alignments"
COUPLING_DIR = os.path.join(ROOT_SAVE, "couplings")
PLOT_DIR = os.path.join(ROOT_SAVE, "plots")
META_DIR = os.path.join(ROOT_SAVE, "meta")
os.makedirs(COUPLING_DIR, exist_ok=True)
os.makedirs(PLOT_DIR, exist_ok=True)
os.makedirs(META_DIR, exist_ok=True)

```

```python
# ----------------------------
# Utilities
# ----------------------------
def seed_everything(seed: int = 42):
    os.environ["PYTHONHASHSEED"] = str(seed)
    random.seed(seed)
    np.random.seed(seed)
    try:
        torch.manual_seed(seed)
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    except Exception:
        pass
    print(f"✅ Seed set to {seed} | Device: {DEVICE}")

def load_all_slices(paths, tps):
    print("Loading AnnData slices...")
    adatas = []
    for i, fh in enumerate(paths):
        a = sc.read_h5ad(fh)
        # Prefer raw counts if available
        if "count" in getattr(a.layers, "keys", lambda: [])():
            a.X = a.layers["count"]
        a.obs["timepoint"] = [tps[i]] * a.shape[0]
        adatas.append(a)
    # Intersect genes across all slices
    common = list(set.intersection(*[set(a.var_names) for a in adatas]))
    adatas = [a[:, common].copy() for a in adatas]
    print(f"Common genes across slices: {len(common)}")
    return adatas

def concat_norm_pca(ad1, ad2, n_comps=PCA_COMP) -> ad.AnnData:
    # Concatenate, normalize, log1p, PCA; then split later by timepoint
    joint = ad.concat([ad1, ad2], join="inner")
    sc.pp.normalize_total(joint)
    sc.pp.log1p(joint)
    sc.pp.pca(joint, n_comps=n_comps)
    return joint

def _labels_or_cluster(adata, key="annotation", n_neighbors=15, resolution=0.8):
    # Use existing annotation if present, else quick Leiden
    if key in adata.obs:
        return adata.obs[key].astype(str).to_numpy()
    tmp = adata.copy()
    sc.pp.neighbors(tmp, n_neighbors=n_neighbors, use_rep="X_pca" if "X_pca" in tmp.obsm else None)
    sc.tl.leiden(tmp, resolution=resolution)
    return tmp.obs["leiden"].astype(str).to_numpy()

def get_feature_views(pre: Dict) -> Tuple[np.ndarray, np.ndarray]:
    """
    Standardize retrieval of feature embeddings for Feature-GW.
    Try common keys; raise a helpful error if unavailable.
    """
    # Common names you might use from mgw_preprocess
    for k1, k2 in [("X_feat", "Z_feat"), ("X", "Z"), ("feat1", "feat2")]:
        if k1 in pre and k2 in pre:
            return np.asarray(pre[k1]), np.asarray(pre[k2])
    raise KeyError(
        "Feature views not found in `pre`. "
        "Expected keys like ('X_feat', 'Z_feat') or ('Z_X','Z_Z'). "
        "Please adapt get_feature_views(pre)."
    )

def normalize_cost(C):
    C = np.asarray(C, dtype=float)
    m = C.max()
    return C / m if m > 0 else C

def coupling_cache_path(pair_key: str, method: str) -> str:
    return os.path.join(COUPLING_DIR, f"{pair_key}__{method}.npz")

def save_dense_as_npz(P: np.ndarray, path: str):
    sp.save_npz(path, sp.csr_matrix(P))

'''
def load_npz_if_exists(path: str):
    return sp.load_npz(path) if os.path.exists(path) else None
'''
def load_npz_if_exists(path: str):
    """
    Load a sparse .npz file if it exists.
    If only a .npy (dense) version is present, load it, convert to sparse, save as .npz, and return the sparse matrix.
    """
    if os.path.exists(path):
        # standard sparse file
        if path.endswith(".npz"):
            return sp.load_npz(path)

        # legacy dense file (.npy)
        if path.endswith(".npy"):
            print(f"[convert] Detected legacy dense file: {path}")
            arr = np.load(path)
            arr = np.asarray(arr, dtype=np.float64)
            spmat = sp.csr_matrix(arr)

            # write back converted version
            new_path = path.replace(".npy", ".npz")
            sp.save_npz(new_path, spmat)
            print(f"[convert] Saved sparse version: {new_path}")

            return spmat
    return None

def procrustes_and_plot(xs, xs2, P, out_png: str, alpha=0.005, lw=0.5):
    try:
        xs_a, xt_a, R, t = plotting.procrustes_from_coupling(xs, xs2, P)
        plotting.plot_alignment_lines_dense(xs_a, xt_a, P, alpha=alpha, lw=lw)
        plt.title("Alignment (Procrustes, lines ∝ coupling)")
        plt.tight_layout()
        plt.savefig(out_png, dpi=200)
        plt.close()
    except Exception as e:
        warnings.warn(f"Plot skipped: {e}")

# ----------------------------
# Method Wrappers (with caching)
# ----------------------------
def run_mgw(pre, tp1: str, tp2: str, **mgw_kwargs) -> np.ndarray:
    # Loading pre-saved MGW couplings
    path = os.path.join(ROOT_SAVE, f"P_{tp1}_{tp2}.npy")
    #path = coupling_cache_path(pair_key, "MGW")
    cached = load_npz_if_exists(path)
    if cached is not None:
        print(f"[cached] MGW: {path}")
        return cached.toarray()
    '''
    print("[run] MGW...")
    out = mgw_mod.mgw_align_core(
        pre,
        widths=mgw_kwargs.get("widths", (128, 256, 256, 128)),
        lr=mgw_kwargs.get("lr", 1e-3),
        niter=mgw_kwargs.get("niter", 20_000),
        knn_k=mgw_kwargs.get("knn_k", KNN_K),
        geodesic_eps=mgw_kwargs.get("geodesic_eps", GEODESIC_EPS),
        save_dir=mgw_kwargs.get("save_dir", os.path.join(ROOT_SAVE, "mgw_temp")),
        tag=pair_key,
        verbose=mgw_kwargs.get("verbose", True),
        plot_net=mgw_kwargs.get("plot_net", True),
        gw_params=mgw_kwargs.get("gw_params", GW_PARAMS),
    )
    P = out["P"]
    save_dense_as_npz(P, path)
    print(f"[saved] MGW → {path}")
    '''
    return P

def run_spatial_gw(xs: np.ndarray, xs2: np.ndarray, pair_key: str) -> np.ndarray:
    path = coupling_cache_path(pair_key, "SpatialGW")
    cached = load_npz_if_exists(path)
    if cached is not None:
        print(f"[cached] Spatial-GW: {path}")
        return cached.toarray()

    print("[run] Spatial-GW...")
    C1 = cdist(xs, xs)**2
    C2 = cdist(xs2, xs2)**2
    C1, C2 = normalize_cost(C1), normalize_cost(C2)
    P = solve_gw_ott(C1, C2, **GW_PARAMS)
    save_dense_as_npz(P, path)
    print(f"[saved] Spatial-GW → {path}")
    return P

def run_feature_gw(Y: np.ndarray, Y2: np.ndarray, pair_key: str) -> np.ndarray:
    path = coupling_cache_path(pair_key, "FeatureGW")
    cached = load_npz_if_exists(path)
    if cached is not None:
        print(f"[cached] Feature-GW: {path}")
        return cached.toarray()

    print("[run] Feature-GW...")
    C1 = cdist(Y, Y)**2
    C2 = cdist(Y2, Y2)**2
    C1, C2 = normalize_cost(C1), normalize_cost(C2)
    P = solve_gw_ott(C1, C2, **GW_PARAMS)
    save_dense_as_npz(P, path)
    print(f"[saved] Feature-GW → {path}")
    return P

from functools import partial

def run_external_with_cache(pair_key: str, method_name: str, fn, *args, **kwargs) -> np.ndarray:
    """
    Run an external method `fn(*args, **kwargs)` that returns a coupling (np.ndarray or sparse),
    with uniform caching to COUPLING_DIR/<pair_key>__<method_name>.npz.
    """
    path = coupling_cache_path(pair_key, method_name)
    cached = load_npz_if_exists(path)
    if cached is not None:
        print(f"[cached] {method_name}: {path}")
        return cached.toarray()

    print(f"[run] {method_name}...")
    P = fn(*args, **kwargs)
    if P is None:
        raise RuntimeError(f"{method_name} returned None")

    if not isinstance(P, np.ndarray):
        # accept scipy sparse or anything with .toarray()
        P = P.toarray() if hasattr(P, "toarray") else np.asarray(P)

    save_dense_as_npz(P, path)
    print(f"[saved] {method_name} → {path}")
    return P

from validation import run_methods as rm
import importlib; importlib.reload(rm)

# Each wrapper returns a dense np.ndarray coupling

def wrap_MOSCOT_translation(st, msi, Z_s, Z_m):
    return rm.run_moscot(st, msi, Z_s, Z_m)

def wrap_SCOT_v2(Z_s, Z_m, k=50, epsilon=1e-3, norm=True):
    return rm.run_scot_v2(Z_s, Z_m, k=k, epsilon=epsilon, norm=norm)

def wrap_SCOT_v1(Z_s, Z_m, k=50, epsilon=1e-3, norm=True):
    return rm.run_scot(Z_s, Z_m, k=k, epsilon=epsilon, norm=norm)
```

```python


def build_competitors_from_processed(A_: ad.AnnData, B_: ad.AnnData, Y: np.ndarray, Y2: np.ndarray):
    """
    Construct the competitor spec dict using the already-processed inputs in evaluate_pair.
    """
    return {
        "MOSCOT_translation": {
            "fn": wrap_MOSCOT_translation,
            "args": [A_, B_, Y, Y2],
        },
        "SCOT_v2": {
            "fn": wrap_SCOT_v2,
            "args": [Y, Y2],
            "kwargs": {"k": 50, "epsilon": 1e-3, "norm": True},
        },
    }

'''
# returns None
"SCOT_v1": {
    "fn": wrap_SCOT_v1,
    "args": [Y, Y2],
    "kwargs": {"k": 50, "epsilon": 1e-3, "norm": True},
},'''

def evaluate_pair(A: ad.AnnData,
                  B: ad.AnnData,
                  tp1: str,
                  tp2: str,
                  cca_comp: int,
                  competitors: Dict[str, dict] | None = None) -> Dict[str, Dict[str, float]]:

    pair_key = f"{tp1}_to_{tp2}"
    print(f"\n=== Pair {pair_key} ===")

    # (unchanged) joint, A_, B_, labels, pre, xs/xs2, Y/Y2 ...
    joint = concat_norm_pca(A, B, n_comps=PCA_COMP)
    A_ = joint[joint.obs["timepoint"] == tp1].copy()
    B_ = joint[joint.obs["timepoint"] == tp2].copy()
    
    A_labels = _labels_or_cluster(A_, key="annotation")
    B_labels = _labels_or_cluster(B_, key="annotation")

    pre = mgw_mod.mgw_preprocess(
        A_, B_,
        PCA_comp=PCA_COMP,
        CCA_comp=cca_comp,
        use_cca_feeler=False,
        use_pca_X=True,
        use_pca_Z=True,
        log1p_X=False,
        log1p_Z=False,
        verbose=True,
        feature_only=True,
        spatial_only=False
    )

    xs  = np.asarray(pre.get("xs",  A_.obsm.get("X_pca")))
    xs2 = np.asarray(pre.get("xs2", B_.obsm.get("X_pca")))
    if xs is None or xs2 is None:
        raise ValueError("Could not find spatial embeddings (`xs`, `xs2`).")

    Y, Y2 = get_feature_views(pre)

    # --- Built-in methods ---
    P_mgw     = run_mgw(pre, tp1, tp2)
    P_spatial = run_spatial_gw(xs, xs2, pair_key)
    P_feature = run_feature_gw(Y, Y2, pair_key)

    # Quick-look plots
    for name, P in [("MGW", P_mgw), ("SpatialGW", P_spatial), ("FeatureGW", P_feature)]:
        out_png = os.path.join(PLOT_DIR, f"{pair_key}__{name}.png")
        procrustes_and_plot(xs, xs2, P, out_png, alpha=0.005, lw=0.5)

    # --- Evaluate all so far ---
    results = {}
    for name, P in [("MGW", P_mgw), ("SpatialGW", P_spatial), ("FeatureGW", P_feature)]:
        results[name] = metrics.evaluate_coupling(xs, xs2, P, A_labels, B_labels, metrics)

    competitors = build_competitors_from_processed(A_, B_, Y, Y2)

    for name, spec in competitors.items():
        fn     = spec["fn"]
        args   = spec.get("args", [])
        kwargs = spec.get("kwargs", {})
        do_plot = spec.get("plot", True)
    
        P = run_external_with_cache(pair_key, name, fn, *args, **kwargs)
    
        # Verify coupling shape matches the processed pair
        if P.shape != (xs.shape[0], xs2.shape[0]):
            raise ValueError(
                f"{name} returned coupling with shape {P.shape}, "
                f"but expected {(xs.shape[0], xs2.shape[0])} matching (A_, B_) after processing."
            )
    
        if do_plot:
            out_png = os.path.join(PLOT_DIR, f"{pair_key}__{name}.png")
            procrustes_and_plot(xs, xs2, P, out_png, alpha=0.005, lw=0.5)
    
        results[name] = metrics.evaluate_coupling(xs, xs2, P, A_labels, B_labels, metrics)
        
    # Persist JSON
    with open(os.path.join(META_DIR, f"{pair_key}__metrics.json"), "w") as f:
        json.dump(results, f, indent=2)

    return results
```

```python
import pandas as pd
from typing import List, Dict
import glob, os

PREFERRED_METHOD_ORDER = [
    "MGW", "SpatialGW", "FeatureGW", "MOSCOT_translation",
    "SCOT_v2", "SCOT_v1"
]

def build_all_pairs_results_from_csv(
    csv_path: str,
    preferred_order: List[str] = PREFERRED_METHOD_ORDER
) -> List[Dict]:
    """
    Rebuild the legacy `all_pairs_results` structure from the tidy CSV.

    Returns a list of dicts:
      {
        "pair": "<tp1>→<tp2>",
        "methods": [method1, method2, ...],
        "metrics": [ {metric: value, ...}, {metric: value, ...}, ... ]  # aligned with methods
      }
    """
    df = pd.read_csv(csv_path)

    # Average duplicates if any (multiple runs of same pair/method)
    agg = df.groupby(["pair", "method"], as_index=False).mean(numeric_only=True)

    # Identify metric columns dynamically
    metric_cols = [c for c in agg.columns if c not in ("pair", "method")]

    all_pairs_results = []
    for pair, sub in agg.groupby("pair", sort=False):
        # Preferred order first, then any remaining methods found in the data
        present = list(sub["method"].unique())
        methods = [m for m in preferred_order if m in present] + [m for m in present if m not in preferred_order]

        metrics_list = []
        for m in methods:
            row = sub[sub["method"] == m]
            if row.empty:
                continue
            d = {k: float(row.iloc[0][k]) for k in metric_cols}
            metrics_list.append(d)

        all_pairs_results.append({
            "pair": pair,
            "methods": methods,
            "metrics": metrics_list
        })

    return all_pairs_results
```
