```python
import matplotlib.pyplot as plt

# Data
data = {
    "MGW": {"mean_cosine": 0.6712460517883301, "median_cosine": 0.6892294883728027, "migration": 0.014474236958333823},
    "Spatial GW": {"mean_cosine": 0.6599059700965881, "median_cosine": 0.6805340051651001, "migration": 0.21839704569435944},
    "Feature GW": {"mean_cosine": 0.7437325119972229, "median_cosine": 0.7531275749206543, "migration": 0.23863534178910376},
    "Moscot": {"mean_cosine": 0.7416129112243652, "median_cosine": 0.7537705898284912, "migration": 0.23366500005513682},
    "SCOT": {"mean_cosine": 0.6648324637583289, "median_cosine": 0.6937531264340652, "migration": 0.23366500005513682},
    "SCOTv2": {"mean_cosine": 0.7137649989732668, "median_cosine": 0.7246853478851697, "migration": 0.24760015609449323},
    "PASTE": {"mean_cosine": 0.4370304088217665, "median_cosine": 0.4551490912678309, "migration": 0.0012193209331708438},
    "POT": {"mean_cosine": 0.7219998256127798, "median_cosine": 0.7384803658123805, "migration": 0.27770939172510484},
}

# Extract arrays
methods = list(data.keys())
x = [data[m]["migration"] for m in methods]
y = [data[m]["mean_cosine"] for m in methods]

# Compute Pareto frontier (maximize y, minimize x)
indices = list(range(len(methods)))
pareto = []
for i in indices:
    xi, yi = x[i], y[i]
    dominated = False
    for j in indices:
        if j == i:
            continue
        xj, yj = x[j], y[j]
        if (xj <= xi and yj >= yi) and (xj < xi or yj > yi):
            dominated = True
            break
    if not dominated:
        pareto.append(i)
pareto_sorted = sorted(pareto, key=lambda i: x[i])

# Plot
plt.figure(figsize=(8, 6))

for i, m in enumerate(methods):
    size = 160 if m.startswith("Us") else 80
    color = "tab:blue" if m.startswith("Us") else "gray"
    plt.scatter(x[i], y[i], s=size, color=color, zorder=3)

# Manually stagger labels to reduce overlap
for i, m in enumerate(methods):
    dx, dy = 0.002, 0.002
    if "SCOTv2" in m: dx, dy = 0.008, -0.01
    elif "Feature" in m: dx, dy = -0.015, 0.004
    elif "Moscot" in m: dx, dy = -0.01, 0.017
    elif "Spatial" in m: dx, dy = 0.01, -0.012
    elif "PASTE" in m: dx, dy = 0.008, 0.01
    elif "POT" in m: dx, dy = -0.004, 0.004
    elif "SCOT" in m: dx, dy = 0.004, 0.008
    elif "Us" in m: dx, dy = 0.005, 0.012

    plt.text(x[i] + dx, y[i] + dy, m, fontsize=15, ha='left', va='bottom')

# Draw Pareto frontier
px = [x[i] for i in pareto_sorted]
py = [y[i] for i in pareto_sorted]
plt.plot(px, py, linestyle="--", color="gray", lw=1)

plt.title("Visium-Xenium Cosine Similarity vs. Spatial Distortion", fontsize=14)
plt.xlabel("Spatial Distortion / Migration (lower is better)", fontsize=12)
plt.ylabel("Mean Cosine Similarity (higher is better)", fontsize=12)
plt.grid(True, linestyle=":", alpha=0.7)
plt.tight_layout()

plt.ylim((0.4,0.78))

out_path = "/home/ph3641/ManifoldAlignment/alignment_scatter.png"
plt.savefig(out_path, dpi=600, bbox_inches="tight")
print("Saved to:", out_path)
```
